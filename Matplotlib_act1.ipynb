```python
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as pd
```

```python

x1 = [10,20,30]
y1 = [20,40,10]
plt.ylabel('Y axis')
x2 = [10,20,30]
y2 = [40,10,30]
plt.xlabel('X axis')

plt.plot(x1, y1, label = "line 1")
plt.plot(x2, y2, label = "line 2")


plt.title('Question 1')
plt.legend()
plt.show()
```

```python
x = [1,4,5,6,7]
plt.xlabel('X axis')

y = [2,6,3,6,3]
plt.ylabel('Y axis')

plt.plot(x, y, color='red', linestyle='dashdot', linewidth=3, marker='o', markerfacecolor='blue', markersize=12)
plt.show()
```

```python
proglang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']

popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(proglang, popularity)
plt.show()
```

```python
proglang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ['yellow', 'green', 'black', 'orange', 'purple', 'blue']

plt.bar(proglang, popularity, color=colors)
plt.show()

```

```python
proglang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']

popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.pie(popularity, labels=proglang, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Popularity of Programming Languages')
plt.axis('equal') 
plt.show()
```
