```python
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4], [1, 2, 3, 1])
plt.ylabel('some numbers')
plt.xlabel('some numbers japon')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1,3,1)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('info')

plt.subplot(1,3,2)
plt.scatter([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('2nd plot')

plt.subplot(1,3,3)
plt.bar([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('3rd plot')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
section = ['A', 'B', 'C', 'D', 'E']
students = [10, 24, 36, 40, 5]
plt.ylabel('number of students')
plt.xlabel('section')
ax.bar(section, students)
#change color of bar into red
ax.bar(section, students, color='black')
plt.show()
```

```python

fig1, ax1 = plt.subplots()

ax1.set_title('number of students in each section')

# Use 'langs' as labels since its length matches 'students'
langs = ['phy', 'chem', 'bio', 'math', 'eng']
ax1.pie(students, labels=langs, autopct='%1.1f%%')
ax1.axis('equal')
plt.show()
```
